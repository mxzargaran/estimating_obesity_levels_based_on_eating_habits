# Visualization of Obesity Levels Data

This report was generated using cleaned data from the [Estimation of Obesity Levels Based On Eating Habits and Physical Condition data set](https://archive.ics.uci.edu/dataset/544/estimation+of+obesity+levels+based+on+eating+habits+and+physical+condition). We plotted each of the feature variables against the different Obesity Levels to try to understand the various data correlation and effects that could help predict obesity. 

## Data Preparation

Import the appropriate libraries, and intake the prepared and cleaned data. Convert the data columns into a readable format before graphing.

```python
# Import
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read cleaned data
df = pd.read_csv("../data/processed/cleaned_data.csv")

print(df.columns)
df.head(5)
```

```python
# Rename columns
df = df.rename(columns={'family_history_with_overweight': 'Overweight_Family_History',
                        'FAVC': 'High_Caloric_Intake', 
                        'FCVC': 'Vegetable_Intake', 
                        'NCP': 'Daily_Meal_Count', 
                        'CAEC': 'Food_Between_Meals', 
                        'SMOKE': 'Smoking', 
                        'CH2O': 'Daily_Water_Intake', 
                        'SCC': 'Monitor_Calories', 
                        'FAF': 'Physical_Activity_Count',
                        'TUE': 'Tech_Device_Usage',
                        'CALC': 'Alcohol_Intake',
                        'MTRANS': 'Transportation',
                        'NObeyesdad': 'Obesity_Level'})

# Obesity Levels Ordering
obesity_values = ["Insufficient_Weight", "Normal_Weight", "Overweight_Level_I", "Overweight_Level_II", "Obesity_Type_I", "Obesity_Type_II", "Obesity_Type_III"]

# Feature Columns
df_categorical_columns = ['Gender', 'Overweight_Family_History', 'High_Caloric_Intake', 'Food_Between_Meals', 'Smoking', 'Monitor_Calories', 'Alcohol_Intake', 'Transportation']
df_numerical_columns = ['Age', 'Height', 'Weight', 'Vegetable_Intake', 'Daily_Meal_Count', 'Daily_Water_Intake', 'Physical_Activity_Count', 'Tech_Device_Usage']
```

## Plot Categorical Features

Visualize the different categorical features and how they are distributed amongst the obesity levels in this data set.

```python
# Categorical Plots
import seaborn as sns

def plot_categorical_features(target, feature, data, order):
    plt.figure(figsize = (13, 5))
    sns.countplot(x=target, hue=feature, data=data, order=order)
    plt.show()

for feature in df_categorical_columns:
    plot_categorical_features('Obesity_Level', feature, df, obesity_values)
```

## Plot Numerical Features

Visualize the different numerical features and the ranges of values they take across the obesity levels in this data set.

```python
# Numerical Plots

def plot_numerical_features(scatter_x, scatter_y):
    plt.figure(figsize = (13, 5))
    plt.scatter(df[scatter_x], df[scatter_y])
    plt.xlabel(scatter_x)
    plt.ylabel(scatter_y)
    plt.tight_layout()
    plt.show()

for num_feature in df_numerical_columns:
    plot_numerical_features('Obesity_Level', num_feature)

```
